# IRI

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data setup (extract a sample portion of your provided dataset for demonstration)
data = 'DataIRI.csv'
data = pd.read_csv(data)

bias = 'Bias.csv'
bias = pd.read_csv(bias)

# Convert to DataFrame
df = pd.DataFrame(data)
bf = pd.DataFrame(bias)
```

```python
# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(bf['BB'], df['V'], color='blue', label='Biological')
plt.scatter(bf['NBB'], df['V'], color='red', label='Non-Biological')

# Add labels and title
plt.title('Scatter Plot: Prospective vs. Psychometric Bias')
plt.xlabel('Psychometric Bias')
plt.ylabel('Total Questionnaire Score')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
